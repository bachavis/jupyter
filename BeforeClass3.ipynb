```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/bchavis1/Downloads/airquality.csv')
```

```python
windNew = df['Wind'].dropna()
stdWind = (windNew-np.mean(windNew)) / np.std(windNew, ddof = 1)
stats.probplot(stdWind, plot=plt)
plt.title("Normal Q-Q Plot: Wind")
plt.ylabel("Observed Quantiles")
plt.show()
```

```python
ozoneNew = df['Ozone'].dropna()
stdOzone = (ozoneNew-np.mean(ozoneNew)) / np.std(ozoneNew, ddof = 1)
stats.probplot(stdOzone, plot=plt)
plt.title("Normal Q-Q Plot: Ozone")
plt.ylabel("Observed Quantiles")
plt.show()
```
